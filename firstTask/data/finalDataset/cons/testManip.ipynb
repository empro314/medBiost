```python
import pandas as pd
import glob
```

```python
fileList = glob.glob('./test/*')
```

```python
fileList
```

```python
dataArray = []
```

```python
for filen in fileList:
    df = pd.read_csv(filen, delimiter='\t', header=None)
    dataArray.append(df)
```

```python
dataArray
```

```python
combinded = dataArray[0]
```

```python
dataArray[1:]
```

```python
for datf in dataArray[1:]:
    combinded = pd.merge(combinded, datf, on=0)
```
