```python
import pandas as pd
import glob as glob
```

```python
geneList = glob.glob("../controls/*/*UQ.txt")
microList = glob.glob("../controls/*/*tion.txt")
```

```python
geneList
```

```python
microList
```

```python
Genes = pd.DataFrame()
Micro = pd.DataFrame()

Genes
```

```python
for el in geneList:
    case = pd.read_csv(el, delimiter='\t', header=None)
    Genes = Genes.append(case.loc[case[0] == 'ENSG00000111653.18'][1], ignore_index=True)
    
case
```

```python
Genes
```

```python
for el in microList:
    case = pd.read_csv(el, delimiter='\t')
    Micro = Micro.append(case.loc[case['miRNA_ID'] == 'hsa-mir-423']['reads_per_million_miRNA_mapped'], ignore_index=True)
    
case
```

```python
Micro
```

```python
Genes.columns = ['exprG']
Micro.columns = ['exprM']
```

```python
Genes
```

```python
Micro
```

```python
ds = pd.concat([Genes, Micro], axis = 1)
```

```python
ds
```

```python
ds.to_csv('controlDataset')
```
